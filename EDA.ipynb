### Picking the output

Initially, we want to predict the ```In-hospital death``` variable. That means keeping it as ```y``` column and dropping other outcome-related variables. These are

* RecordID
* SAPS-I score
* SOFA score
* Length of stay (days)
* Survival (days)
* In-hospital death (0: survivor, or 1: died in-hospital)


````python
drop_cols = ["recordid", "SAPS-I", "SOFA", "Length_of_stay", "Survival"]
y_col = "In-hospital_death"

X, y = df.drop(columns = drop_cols + [y_col]), df[y_col]
````

````python
X_PATH = "./data/Train.csv"
y_PATH = "./data/y.csv"

X.to_csv(X_PATH, index=False)
y.to_csv(y_PATH, index=False)
````
